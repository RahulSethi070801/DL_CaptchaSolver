```python
import pickle
from keras.models import Sequential
from keras.layers import Dense,Dropout, Activation,Flatten
from keras.layers import Conv2D, MaxPooling2D
import matplotlib.pyplot as plt
import numpy as np
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
```

```python
def loadPickleData(file):
    data = pickle.load(open(file, 'rb'));
    images = data[0];
    labels = data[1];
    return [images, labels]
```

```python
[images,labels] = loadPickleData("./../Data/pickledFiles/Num_SingleDigit.p")
```

```python
y = np_utils.to_categorical(labels);
print(y);
print(y.shape)
```

```python
X = images
```

```python
X.shape
```

```python
X = np.reshape(X, (2000,60,40,1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1);
```

```python
model = Sequential();
```

```python
model.add(Conv2D(20,(2,2),strides = 1, activation = 'relu', input_shape = (60,40,1)))
model.add(MaxPooling2D(pool_size = (2,2)));
```

```python
model.add(Conv2D(40,(3,3),strides = 1, activation = 'relu'))
model.add(Dropout(0.1));
model.add(Conv2D(80,(2,2),strides = 1, activation = 'relu'))
```

```python
model.add(Conv2D(40,(2,2),strides = 1, activation = 'relu'))
```

```python
model.add(MaxPooling2D(pool_size = (2,2)));
```

```python
model.add(Flatten())
```

```python
model.add(Dense(500, activation = 'relu'))
model.add(Dropout(0.1));
model.add(Dense(200, activation = 'relu'))
model.add(Dropout(0.1));
```

```python
num_classes = y.shape[1]
model.add(Dense(num_classes, activation = 'softmax'));
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=1000, epochs=500, verbose=1);
```

```python
print(model.evaluate(X_test, y_test))
```
