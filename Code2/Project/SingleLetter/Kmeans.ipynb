```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import pickle
```

```python
def loadPickleData(file):
    data = pickle.load(open(file, 'rb'));
    images = data[0];
    labels = data[1];
    return [images, labels]
```

```python
[images,labels] = loadPickleData("./../Data/pickledFiles/Num_SingleDigit.p")
```

```python
nTrainPrcntg = 1
nTrain = int(nTrainPrcntg*images.shape[0])
trainData0 = images[0:nTrain,:].reshape((2000, -1))
trainLabel = labels[0:nTrain,]
testData0 = images[nTrain:,:].reshape((2000, -1))
testLabel = labels[nTrain:,]
```

```python
trainData0.shape
```

```python
PCAcontrol = 1
if PCAcontrol == 1:
    pca = PCA()
    trainData = pca.fit_transform(trainData0)
else:
    trainData = trainData0
    testData = testData0
```

```python
kmeans = KMeans(n_clusters = 10, verbose = 1)
kmeans.fit(trainData)
trainPredict = kmeans.predict(trainData)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(trainPredict, trainLabel)
```
