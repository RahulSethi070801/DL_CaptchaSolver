```python
from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
```

```python
dir = 'Dataset/Alpha/Len3Char/'
outdir = 'Dataset/Alpha/Len3Char/'
```

```python
## if outdir == dir, then the code will overwrite the original database, which is fine...save some space
# folders = ['Num', 'Alpha', 'AlphaNum']
folders = ['Alpha']
val = [5]
for fold in folders:
    for value in val:
        print(fold, value)
        dir = 'Dataset/' + str(fold) + '/Len' + str(value) + 'Char/'
        outdir = 'Dataset/' + str(fold) + '/Len' + str(value) + 'Char/'
        # counter = 0;
        for image in tqdm(os.listdir(dir)):
            if (dir+image == 'Dataset/Alpha/Len5Char/EIZXE_1.png'):
                continue
            img = Image.open(dir+image);
            # print(type(img));
            img2 = img.crop((0, 0, 40*value,60))
            # plt.figure()
            # plt.imshow(img2)
            img2.save(outdir+image)
            # counter+=1;
            # print(counter)
            # if(counter > 10):
            #     break;
        plt.show()
```
